## Исследование модели
Нам дана обученная **модель сходства (Similarity model)**. Модели такого типа получают на вход два изображения и определяют, насколько эти изображения схожи между собой. Основанный на сходстве подход к распознаванию лиц позволяет создавать более эффективные системы, чем подход, основанный на классификации, так как алгоритм сходства более устойчив по отношению к изменению групп классификации.

Описание содержимого исходного utils.py:
1) `__forward(model: Model, img: np.ndarray)` - возвращает предсказание модели,
2) `compare(model: Model, img1_path: str, img2_path: str)` - сравнивает два изображения при помощи `__calculate_embedding(model: Model, img_path: str)` и `__calculate_cosine(emb1, emb2)`,
3) `__find_face(img_path)` - находит лицо на изображении,
4) `__resize(img, target_size)` - приводит изображение к нужному формату,
5) `__calculate_embedding(model: Model, img_path: str)` - при помощи `__find_face(img_path)` находит лицо на изображении, приводит к нужному формату (`__resize(img, target_size)`) и возвращает векторное представление из предсказания модели (`__forward(model: Model, img: np.ndarray)`),
6) `__calculate_cosine(emb1, emb2)` - находит косинусное расстояние между векторами.

### Вспомогательный этап - сохранение на диск

#### Сохраним векторные представления изображений из id_samples на диск для ускорения работы модели

```python
filelist=os.listdir('id_samples')
for i in range(len(filelist)):  
    c = __calculate_embedding(face_model, "id_samples/"+filelist[i]) 
    path = 'id_samples_arrays/' + filelist[i][:-4]+ '.npy'
    np.save(path, c)
print('done')
```

#### Сохраним в текстовый файл предсказания модели

```python
def compare_emb(model: Model, img1_path: str, img2_path: str) -> Dict[str, Any]:
    emb1 = np.load(img1_path)
    emb2 = np.load(img2_path)
    distance = __calculate_cosine(emb1, emb2)
    return {"confidence": 1.0, "distance": distance, "equal": distance <= 0.4}
```

```python
filelist=os.listdir('id_samples_arrays')
eq_dict = {}
f = open('id_samples_emb_equal.txt', 'w')
for ind1 in range(len(filelist)):
    for ind2 in range(ind1+1, len(filelist)):
        c = compare_emb(face_model, "id_samples_arrays/"+filelist[ind1], "id_samples_arrays/"+ filelist[ind2]) 
        eq_dict[(filelist[ind1], filelist[ind2])] = c
        f.write(filelist[ind1] + '\t' + filelist[ind2] + '\t' + str(c['distance']) + '\n')
print('done')
f.close()
```

### Оценка качества модели

Перед тем, как начать сбор **OOD набора данных**, оценим качество исследуемой модели, а именно посчитаем:
1) `True Positive Rate`,
2) `False Positive Rate`,
3) `Presicion`,
4) `Recall`.
   
Построим:
1) `Receiver Operating Characteristic (ROC) Curve`
2) `Precision-Recall (PR) Curve`
3) `Confusion Matrix`

#### Отобразим результаты сравнений 

```python
df = pd.read_csv('id_samples_emb_equal.txt', sep='\t', header=None)

df.columns = ['photo1', 'photo2', 'pred']

df['photo1'] = df['photo1'].apply(lambda x: x.rsplit('.', 1)[0])
df['photo2'] = df['photo2'].apply(lambda x: x.rsplit('.', 1)[0])

df['name1'] = df['photo1'].apply(lambda x: '_'.join(x.split('_')[:-1]))
df['name2'] = df['photo2'].apply(lambda x: '_'.join(x.split('_')[:-1]))

df['label'] = df.apply(lambda row: 1 if row['name1'] == row['name2'] else 0, axis=1)

df = df.drop(columns=['name1', 'name2'])

df
```

#### Медианное, среднее значения выборки, а так же сдвиг (так как метрикой расстояния являетя косинусное расстояние, а не косинусная близость)

```python
df['pred'] = 1 - df['pred']
```

```python
df[df['label'] == 1].pred.median()
```

```python
df[df['label'] == 0].pred.mean()
```

#### Риусем ROC кривую

```python
fpr, tpr, thresholds = roc_curve(df['label'], df['pred'])

roc_auc = roc_auc_score(df['label'], df['pred'])

plt.plot(fpr, tpr, color='blue', label=f'ROC Curve (AUC = {roc_auc:.4f})')
plt.plot([0, 1], [0, 1], color='red', linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.grid()
plt.show()


print(f'ROC AUC: {roc_auc:.4f}')
```

#### True Positive Rate, False Positive Rate

```python
fpr, tpr, thresholds = roc_curve(df['label'], df['pred'])

desired_fpr = 0.1 

optimal_idx = np.where(fpr <= desired_fpr)[0]  
optimal_threshold_idx = optimal_idx[np.argmax(tpr[optimal_idx])]  
optimal_threshold = thresholds[optimal_threshold_idx]

print(f'Optimal Threshold: {optimal_threshold:.4f}')
print(f'TPR at Optimal Threshold: {tpr[optimal_threshold_idx]:.4f}')
print(f'FPR at Optimal Threshold: {fpr[optimal_threshold_idx]:.4f}')
```

#### Строим Presicion-Recall кривую

```python
precision, recall, thresholds = precision_recall_curve(df['label'], df['pred'])

average_precision = average_precision_score(df['label'], df['pred'])

plt.figure(figsize=(8, 6))
plt.plot(recall, precision, color='blue', label=f'PR Curve (AP = {average_precision:.4f})')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall (PR) Curve')
plt.legend(loc='upper right')
plt.grid()
plt.show()

print(f'Average Precision Score: {average_precision:.4f}')
```

#### Строим Confusion matrix

```python
threshold = 0.6
y_pred_class = df['pred'] > threshold

cm = confusion_matrix(df['label'], y_pred_class)

cm_normalized = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]

plt.figure(figsize=(6, 4))
sns.heatmap(cm_normalized, annot=True, fmt='.2f', cmap='Blues',
            xticklabels=['Predicted Negative', 'Predicted Positive'],
            yticklabels=['Actual Negative', 'Actual Positive'])
plt.title('Normalized Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```

### Проверим условие задачи

В OOD наборе данных должны содержаться не менее 50 примеров, для которых хотя бы один пример из предоставленного ID набора определялся, как близкий.

```python
def compare_emb_img(model: Model, img1_path: str, img2_path: str) -> Dict[str, Any]:
    emb1 = np.load(img1_path)
    emb2 = __calculate_embedding(model, img2_path)
    distance = __calculate_cosine(emb1, emb2)
    return {"confidence": 1.0, "distance": distance, "equal": distance <= 0.4}
```

```python
k = 0
filelist_ID=os.listdir('id_samples_arrays')
filelist_OOD=os.listdir('OOD')
for i in range(len(filelist_OOD)):
    for j in range(len(filelist_ID)):
        c = compare_emb_img(face_model, "id_samples_arrays/"+filelist_ID[j], "OOD/"+ filelist_OOD[i]) 
        if c['equal']:
            k += 1
            print(f'{k}) Similar')
            print(filelist_OOD[i], filelist_ID[j], c)
            break
    

```
